```python
# ============================================================
# Cu-64 and N-13 Yield Plot
# Dual X-axis:
#   Bottom : Incident proton energy
#   Top    : Mean proton energy after Havar foil
#
# Y-axis:
#   Yield (nuclei/proton)
#
# Author: Your Name
# ============================================================

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

```

```python
# ============================================================
# USER INPUT
# ============================================================

# Incident proton energies (MeV)
incident_energy = np.array([
    11, 12, 13, 14, 15, 16
])

# Mean proton energy after Havar foil (MeV)
after_havar_energy = np.array([
    10.3, 11.2, 12.1, 13.0, 13.9, 14.8
])

# Cu-64 yields (nuclei/proton)
cu64_yield = np.array([
    3.2e-5,
    4.1e-5,
    5.5e-5,
    6.2e-5,
    5.8e-5,
    4.9e-5
])

# N-13 yields (nuclei/proton)
n13_yield = np.array([
    2.8e-4,
    3.1e-4,
    3.5e-4,
    3.9e-4,
    4.3e-4,
    4.8e-4
])

```

```python
# ============================================================
# CREATE FIGURE
# ============================================================

fig, ax = plt.subplots(figsize=(10, 6))

# ============================================================
# MAIN PLOT
# ============================================================

ax.plot(
    incident_energy,
    cu64_yield,
    marker='o',
    markersize=8,
    linewidth=2.5,
    label=r'$^{64}$Cu'
)

ax.plot(
    incident_energy,
    n13_yield,
    marker='s',
    markersize=8,
    linewidth=2.5,
    label=r'$^{13}$N'
)

# ============================================================
# Y-AXIS
# ============================================================

ax.set_ylabel(
    r'Yield (nuclei proton$^{-1}$)',
    fontsize=15
)

# ============================================================
# BOTTOM X-AXIS
# ============================================================

ax.set_xlabel(
    'Incident Proton Energy (MeV)',
    fontsize=15
)

ax.set_xticks(incident_energy)

# ============================================================
# TOP X-AXIS
# ============================================================

ax_top = ax.twiny()

ax_top.set_xlim(ax.get_xlim())

ax_top.set_xticks(incident_energy)

ax_top.set_xticklabels(
    [f'{x:.1f}' for x in after_havar_energy]
)

ax_top.set_xlabel(
    'Mean Proton Energy After Havar Foil (MeV)',
    fontsize=15
)

# ============================================================
# STYLE
# ============================================================

ax.grid(
    True,
    linestyle='--',
    alpha=0.4
)

ax.tick_params(
    axis='both',
    labelsize=12
)

ax_top.tick_params(
    axis='x',
    labelsize=12
)

# Scientific notation on Y-axis
ax.ticklabel_format(
    axis='y',
    style='sci',
    scilimits=(0, 0)
)

# ============================================================
# LEGEND
# ============================================================

ax.legend(
    fontsize=13,
    frameon=True
)

# ============================================================
# TITLE
# ============================================================

plt.title(
    r'Production Yield of $^{64}$Cu and $^{13}$N',
    fontsize=17,
    pad=18
)

# ============================================================
# LAYOUT
# ============================================================

plt.tight_layout()

# ============================================================
# SAVE FIGURE
# ============================================================

plt.savefig(
    'Cu64_N13_yield.png',
    dpi=600,
    bbox_inches='tight'
)

# ============================================================
# SHOW
# ============================================================

plt.show()
```
